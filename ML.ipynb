```python
from sklearn.datasets import load_digits

digits = load_digits()
#digits.DESCR
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(digits.data,digits.target,random_state=10,test_size=0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X=x_train,y=y_train)
```

```python
predicted=knn.predict(X=x_test)
expected=y_test
```

```python
predicted[:30]
```

```python
expected[:30]
```

```python
print(f"{knn.score(x_test,y_test):.2f}")
```

```python
from sklearn.metrics import confusion_matrix
con = confusion_matrix(y_true=expected,y_pred=predicted)
con
```

```python
from sklearn.metrics import classification_report
names =[str(digits) for digits in digits.target_names]
names
```

```python
print(classification_report(expected,predicted,target_names=names))
```
